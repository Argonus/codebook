```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

# Nawiązanie połączenia z bazą danych
base_connection = "DRIVER={ODBC Driver 17 for SQL Server};SERVER=analityk.wwsi.edu.pl,50221;uid=student;pwd=ciekawski"
synop_db = f"{base_connection};DATABASE=synop"
connection_url = URL.create("mssql+pyodbc", query={"odbc_connect": synop_db})
engine = create_engine(connection_url)
```

```python
# Data selection and verification
base_query = """
select Nazwa,Wojewodztwo,Szerokosc,Dlugosc,
       Data,godzina,
       WysokoscPodstawyChmur,Widzialnosc,WielkoscZachmurzenia,
       KierunekWiatru,PredkoscWiatru,
       TemperaturaPowietrza,TemperaturaPunktuRosy,CisnienieNaPoziomieMorza,
       WysokoscOpadu,OkresPomiaruOpadu,
       CzyDeszcz,CzyBurza,CzyMgla,CzyOpadMarznacy,
       CzyPrzelotny,CzySnieg,CzyZamiec
from Depesze as D join stacje as S
                       on d.stacja=s.idstacji
where Data between '20170101' and '20201231'
"""
results = None
with engine.begin() as conn:
    results = pd.read_sql_query(sa.text(base_query), conn)
```

```python
# Filter By Long/Lat
# Zakresy geograficzne dla Polski, by bounding box
MIN_LATITUDE = 49.0273953314
MAX_LATITUDE = 54.8515359564
MIN_LONGITUDE = 14.0745211117
MAX_LONGITUDE = 24.0299857927

init_size = len(results)
df = results.copy()

# Konwersja kolumn na liczby, zamiana błędnych wartości na NaN
df['Szerokosc'] = pd.to_numeric(df['Szerokosc'], errors='coerce')
df['Dlugosc'] = pd.to_numeric(df['Dlugosc'], errors='coerce')

# Usuwanie wierszy z NaN (błędy konwersji lub brakujące dane)
df = df.dropna(subset=['Szerokosc', 'Dlugosc'])

# Filtrowanie danych według zakresu
filtered_df = df[
    (df['Szerokosc'] >= MIN_LATITUDE) &
    (df['Szerokosc'] <= MAX_LATITUDE) &
    (df['Dlugosc'] >= MIN_LONGITUDE) &
    (df['Dlugosc'] <= MAX_LONGITUDE)
]

filtered_size = len(filtered_df)
result = init_size - filtered_size

print(result, "records removed")
print(filtered_size, "records left")

# We can find real long/lat and updates base on station name
```

```python
df = filtered_df.copy()
# Filter by WysokoscPodstawyChmur
MIN_CLOUD_BASE = 0    # Chmury mogą zaczynać się od poziomu gruntu
MAX_CLOUD_BASE = 12000  # Górna granica dla podstawy chmur
df['WysokoscPodstawyChmur'] = pd.to_numeric(df['WysokoscPodstawyChmur'], errors='coerce')

filtered_by_wpc = df[
    (df['WysokoscPodstawyChmur'] >= MIN_CLOUD_BASE) &
    (df['WysokoscPodstawyChmur'] <= MAX_CLOUD_BASE)
]

init_size = len(filtered_df)
filtered_size = len(filtered_by_wpc)
result = init_size - filtered_size

print(result, "records removed")
print(filtered_size, "records left")
```

```python
# Filter by visibility
df = filtered_by_wpc.copy()

# Zakresy widzialności (w metrach)
MIN_VISIBILITY = 0       # Widzialność nie może być ujemna
MAX_VISIBILITY = 100000  # Górna granica (100 km)

# Konwersja kolumny na liczbę
df['Widzialnosc'] = pd.to_numeric(df['Widzialnosc'], errors='coerce')

# Usuwanie wierszy z wartościami spoza zakresu
filtered_by_vis = df[
    (df['Widzialnosc'] >= MIN_VISIBILITY) &
    (df['Widzialnosc'] <= MAX_VISIBILITY)
]

# Wyświetlenie przefiltrowanych danych
init_size = len(df)
filtered_size = len(filtered_by_vis)
result = init_size - filtered_size

print(result, "records removed")
print(filtered_size, "records left")
```

```python
df['Rok'] = df['Data'].dt.year
df['Miesiac'] = df['Data'].dt.month

# Calculations
monthly_temperature = df.groupby(['Rok', 'Miesiac'])['TemperaturaPowietrza'].mean().unstack(level=0)

# Wykres średnich miesięcznych temperatur
plt.figure(figsize=(12, 6))
monthly_temperature.plot(marker='o')
plt.title('Średnia miesięczna temperatura powietrza dla każdego roku')
plt.xlabel('Miesiąc')
plt.ylabel('Średnia temperatura (°C)')
plt.legend(title='Rok')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

monthly_temperature.head()
```

```python
# Usunięcie dnia przestępnego (29 lutego)
df = df[~((df['Data'].dt.month == 2) & (df['Data'].dt.day == 29))]

daily_amplitudes = (df.groupby(['Rok', 'DzienRoku'])['TemperaturaPowietrza'].agg(['min', 'max']).reset_index())
daily_amplitudes['Amplituda'] = daily_amplitudes['max'] - daily_amplitudes['min']
daily_amplitudes_pivot = daily_amplitudes.pivot(index='DzienRoku', columns='Rok', values='Amplituda')

# Wykres dziennej amplitudy temperatury na przestrzeni roku dla każdego roku
plt.figure(figsize=(14, 7))
for year in daily_amplitudes_pivot.columns:
    plt.plot(daily_amplitudes_pivot.index, daily_amplitudes_pivot[year], label=year)

plt.title('Dzienna amplituda temperatury powietrza na przestrzeni roku (2017–2020, bez 29 lutego)')
plt.xlabel('Dzień roku')
plt.ylabel('Amplituda temperatury (°C)')
plt.legend(title='Rok', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
